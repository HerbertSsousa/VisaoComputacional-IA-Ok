```python
# Importar o SDK de infer√™ncia e as bibliotecas necess√°rias
from inference_sdk import InferenceHTTPClient
from google.colab import files
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt
import requests

# para analisar a imagem, baixe e escolha os arquivos que deseja procurar carros!


# Inicializar o cliente com a URL e a chave da API
CLIENT = InferenceHTTPClient(
    api_url="https://detect.roboflow.com",
    api_key="PH63YsNXfScM9AgIGGR5"
)

# Fazer o upload da imagem
uploaded = files.upload()
image_path = list(uploaded.keys())[0]

# Fazer a infer√™ncia na imagem enviada
result = CLIENT.infer(image_path, model_id="modelo-j8zfk/1")
print(result)

# Abrir a imagem carregada
image = Image.open(image_path)

# Criar um objeto de desenho
draw = ImageDraw.Draw(image)

# Se o resultado retornar previs√µes com caixas delimitadoras
if 'predictions' in result:
    for prediction in result['predictions']:
        # Obter as coordenadas da caixa delimitadora
        x = prediction['x']
        y = prediction['y']
        width = prediction['width']
        height = prediction['height']
        left = x - width / 2
        top = y - height / 2
        right = x + width / 2
        bottom = y + height / 2

        # Desenhar a caixa delimitadora na imagem
        draw.rectangle([left, top, right, bottom], outline="red", width=3)

        # Adicionar o label (nome do objeto) acima da caixa
        label = prediction['class']
        draw.text((left, top - 10), label, fill="red")

# Exibir a imagem com as previs√µes usando matplotlib
plt.figure(figsize=(8, 8))
plt.imshow(image)
plt.axis('off')
plt.show()
```
